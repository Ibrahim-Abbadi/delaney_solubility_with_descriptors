### Load data

```python
import pandas as pd
```

```python
df= pd.read_csv('delaney_solubility_with_descriptors.csv')
df
```

### data preperation

```python
y=df['logS']
y
```

```python
X=df.drop('logS',axis =1)
X
```

### Data spleating

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
X_train 
```

## Model building 

### Linear regression 

#### training the model 

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

#### Make a prediction 

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_test_pred
```

```python
y_lr_train_pred
```

#### Model performance 

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)


lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['LinearRegression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = (['Method', 'Training MSE', 'training R2', 'testing MSE', 'testing R2'])
lr_results
```

### Random forest

#### training the model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth = 2, random_state = 100)
rf.fit(X_train, y_train)
```

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train,y_lr_train_pred)
rf_train_r2 = r2_score(y_train, y_lr_train_pred)


rf_test_mse = mean_squared_error(y_test, y_lr_test_pred)
rf_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = (['Method', 'Training MSE', 'training R2', 'testing MSE', 'testing R2'])
rf_results
```

### Model comparison 

```python
df_models = pd.concat([lr_results, rf_results], axis = 0)
df_models
```

```python
df_models.reset_index(drop = True)
```
